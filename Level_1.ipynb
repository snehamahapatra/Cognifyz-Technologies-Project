### Task 3
#### Geospatial Analysis
- Visualize the locations of restaurants on a map using latitude and longitude information.
-Analyze the distribution of restaurants across different cities or countries.
- Distribution if there is any correlation between the restaurant's location and its rating.

```python
import folium
```

```python
#calculate the map center
m = folium.Map(location=[data['Latitude'].mean(), data['Longitude'].mean()], zoom_start=10)
```

```python
for _, row in data.iterrows():
    folium.Marker([row['Latitude'], row['Longitude']], popup=row['Restaurant Name']).add_to(m)
```

```python
m.save("restaurant_map.html")
```

```python
city_distribution = data['City'].value_counts()
```

```python
plt.figure(figsize=(12, 6))
city_distribution.plot(kind='bar')
plt.title('Restaurant Distribution by City')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.show()
```

```python
correlation_latitude_rating = data['Latitude'].corr(data['Aggregate rating'])
correlation_longitude_rating = data['Longitude'].corr(data['Aggregate rating'])
```

```python
print(f"Correlation between Latitude and Rating: {correlation_latitude_rating}")
print(f"Correlation between Longitude and Rating: {correlation_longitude_rating}")
```

```python
# Analyze the distribution of restaurants across different cities or countries.

city_counts = data['City'].value_counts()

# Count the number of restaurants in each country
country_counts = data['Country Code'].value_counts()

# Plot distribution by city
plt.figure(figsize=(12, 6))
city_counts.head(20).plot(kind='bar')
plt.title('Restaurant Distribution by City')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.show()
```

```python
# Plot distribution by country
plt.figure(figsize=(12, 6))
country_counts.head(20).plot(kind='bar')
plt.title('Restaurant Distribution by Country')
plt.xlabel('Country')
plt.ylabel('Number of Restaurants')
plt.show()
```
